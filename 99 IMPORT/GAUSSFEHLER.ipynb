```python
import matplotlib
import matplotlib.pyplot as plt
from scipy.special import erfc, erf
import numpy as np
```

```python
# Data for plotting
t = np.arange(-3, 3, 0.01)
s1 = erf(t)
s2 = erfc(t)
```

```python
fig, ax = plt.subplots()
ax.plot(t,s1)
ax.plot(t,s2)

ax.set(xlabel='x', ylabel='erf(x)/erfc(x)',
       title='Gauss Fehler Funktion')
ax.grid()

fig.savefig("test.png")
plt.show()
```
